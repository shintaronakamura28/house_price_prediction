```python
def explore_data(df):
    num_rows = df.shape[0]
    num_cols = df.shape[1]
    data_types = df.dtypes
    
    print(f'Number of rows:',{num_rows})
    print(f'Number of columns:', {num_cols})
    print(f'\nData types:', data_types)  

explore_data(df)
```

```python
def perform_eda(df):
    duplicates = df.duplicated().sum()
    miss_vals = df.isna().sum()

    print(f'Number of Duplicates: {duplicates}')
    print(f'\nNumber of Missing values: \n{miss_vals}')
    print('\nStatistics')

perform_eda(df)
```

```python
def check_col_values(df):
    print('Check all column values for anomalies \n')

    for column in df.columns:
        print(f'Unique Valies in {column}')
        print(df[column].unique())
        print('\n')

check_col_values(df)
```

```python
def plot_correlation_heatmap(df):
    fig, ax = plt.subplots(figsize= (15,5))

    #define the mask to set the values in the upper traingle to false
    mask = np.triu(np.ones_like(df.corr(), dtype=np.bool))

    #plot the heatmap
    sns.heatmap(
        df.corr().round(2),
        ax=ax,
        mask=mask,
        vmin=-1,
        vmax=1,
        annot=True,
        cmap='BrBG'
    )

    ax.set_title('Triangle Correlation Heatmap', fontdict={'fontsize': 20}, pad=16)
    plt.show()

plot_correlation_heatmap(df)
```

```python
# Correlation heatmap as ranked
def ranked_correlation_heatmap(df, target_column):
    plt.figure(figsize= (8,8))
    heatmap = sns.heatmap(
        df.corr()[[target_column]].sort_values(by=target_column, ascending=False),
        vmin= -1, 
        vmax = 1, 
        annot = True, 
        cmap= 'BrBG'
    )
    
    heatmap.set_title(f'Features Correlation with {target_column}', fontdict={'fontsize': 12}, pad=10)
    plt.show()

ranked_correlation_heatmap(df, target_column='SalePrice')
```

```python
def visualize_data(df, column_name):
    fig, axes = plt.subplots(nrows=2, sharex=False, figsize=(8,6))
    sns.set_style('dark')


    #histogram plot
    sns.histplot(df[column_name], ax=axes[0], bins='auto', kde=True)
    axes[0].tick_params(axis='x', labelrotation=45)

    #boxplot
    sns.boxplot(data=df, x=column_name, ax=axes[1])
    plt.tight_layout()
    plt.show
visualize_data(df,  column_name='SalePrice')
```
